```python
path = r'C:\Users\liuzi\Desktop\Jupiter PDF'
files = os.listdir(path)
fundname = list()

for file in files:
    if file.endswith(".pdf"):
        name = re.findall(r".+(?=Factsheet)", file)
        # name = re.findall(r"/jupiter.*?(?<=fund)", file)
        if len(name) > 0:
            fundname.append(name[0])
        elif len(name) == 0:
            name = re.findall(r"(?<=NCO-).*?fund", file)
            if len(name)>0:
                fundname.append(name[0])
            elif len(name) ==0:
                 pdf=pdfplumber.open(file)
                 pages=pdf.pages
                 firstpage = pdf.pages[0]
                 name = firstpage.extract_text()
                 name = name[name.find('Jupiter'):name.find('Fund')+4]
                 fundname.append(name)


print(fundname)
print(len(fundname))
```

```python

path = r'C:\Users\liuzi\Desktop\Jupiter PDF'
files = glob(path + "/*.pdf")

lst1m = []
lst3m = []
lstYTD = []
lst1yr = []
lst3yrs = []
lst5yrs = []
lst10yrs = []
AuM = []
PDFname = []
currency = []

for file in files:
    print(file)
    pdf=pdfplumber.open(file)
    pages=pdf.pages
    firstpage = pdf.pages[0]
#################查找 排名 quarter ranking#############################
    tab = firstpage.extract_text()
    tab = tab.replace(' ','')
    locator = tab.find('QuartileRanking')+ 15 # 15 是quartile ranking 字符串的长度
    Qranking = tab[locator:locator+7]
    Qranking = re.sub('[a-zA-Z]', '-',Qranking)# 替换掉找出来的乱码
    
    lst1m.append(Qranking[0])
    lst3m.append(Qranking[1])
    lstYTD.append(Qranking[2])
    lst1yr.append(Qranking[3])
    lst3yrs.append(Qranking[4])
    lst5yrs.append(Qranking[5])
    lst10yrs.append(Qranking[6])
#################查找 AuM #############################
    secondpage = pdf.pages[1]
    aum = secondpage.extract_text()
    aum = aum.replace(' ','')
    locator2 = aum.find('FundValue')+ 9 # 9 是fund value 字符串的长度
    # print(locator2)
    aumNo = aum[locator2:locator2+10]
    aumNo = aumNo.replace(',','')
    No = re.findall('\d+', aumNo)
    if len(No)> 0:
        No = float(No[0])
    else:
        No = ''
    AuM.append(No)
################查找 AuM Currency #############################
    if aumNo.find('USD')>0:
        currency.append('USD')
    elif aumNo.find('GBP')>0:
        currency.append('GBP')
    elif aumNo.find('EUR')>0:
        currency.append('EUR')
    elif aumNo.find('CHF')>0:
        currency.append('CHF')
    elif aumNo.find('£')>0:
        currency.append('£')
    elif aumNo.find('$')>0:
        currency.append('£')
    else:
        currency.append('-')
################# 附上每个文件名字，方便查找问题 #############################
    PDFname.append(file)
```
